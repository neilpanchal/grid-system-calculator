```python
def isratio(gdf, ratio):
    return np.isclose(ratio, gdf.values[ :, None], 
           rtol=1e-05, atol=1e-08, equal_nan=False).sum()

result = pd.DataFrame(columns=['WIDTH', 'HEIGHT', 'SQUARE', '35MM', 'TV','HD', 'WIDE'])

# for iw in range(492,589):
#     for ih in range(648, 769):

# A4_width = 595.276
# A4_height = 841.89

# w_max = US_Letter_width - 72
# w_min = US_Letter_width - 120

# h_max = US_Letter_height - 72
# h_min = US_Letter_height - 180

w_max = int(A4_width - 71.276)
w_min = int(A4_width - 119.276)

h_max = int(A4_height - 71.89)
h_min = int(A4_height - 179.89)


for iw in range(w_min,w_max):
    for ih in range(h_min, h_max):
        if ((ih % 12) == 0 ):
            dfl = []
            for col in range(1,17):
                for row in range(1,17):
                    dfl.append({'COL': col, 'ROW': row})

            df = pd.DataFrame(dfl)
            df['C'] = (iw - col_gutter * (df['COL']-1)) / df['COL']
            df['R'] = (ih - row_gutter * (df['ROW']-1)) / df['ROW']
            df['RC'] = df['R'] / df['C']
            df['CR'] =  df['C'] / df['R']
            df['RG-C'] = (df['R'] + row_gutter) / df['C']
            df['C-RG'] = df['C'] / (df['R'] + row_gutter)
            df['R-CG'] = df['R'] / (df['C'] + col_gutter)
            df['CG-R'] = (df['C'] + col_gutter) / df['R']
            df['RG-CG'] = (df['R'] + row_gutter) / (df['C'] + col_gutter)
            df['CG-RG'] = (df['C'] + col_gutter) / (df['R'] + row_gutter)

            square_sum = df.iloc[:, 4:].apply(isratio, args=([ratio_square],)).sum().sum()
            mm_sum = df.iloc[:, 4:].apply(isratio, args=([ratio_35mm, 1/ratio_35mm],)).sum().sum()
            tv_sum = df.iloc[:, 4:].apply(isratio, args=([ratio_tv, 1/ratio_tv],)).sum().sum()
            hd_sum = df.iloc[:, 4:].apply(isratio, args=([ratio_hd1, ratio_hd2, 1/ratio_hd1, 1/ratio_hd2],)).sum().sum()
            wide_sum = df.iloc[:, 4:].apply(isratio, args=([ratio_wide, ratio_ultrawide, 1/ratio_wide, 1/ratio_ultrawide],)).sum().sum()

            result = result.append({'WIDTH': iw, 'HEIGHT': ih, 
                                    'SQUARE': square_sum,
                                    '35MM': mm_sum,
                                    'TV': tv_sum,
                                    'HD': hd_sum,
                                    'WIDE': wide_sum,
                                   }, ignore_index=True)
    print('.', end='')
```
